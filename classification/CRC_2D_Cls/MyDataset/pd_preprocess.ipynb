```python
import os
import numpy as np
import pandas as pd
def create_df_cls(path):
    count = 0
    # col_name = ['img_name', 'img_path', 'img_label']
    col_name = ['img_dir','img_name', 'img_path']
    col_name = ['img_dir','img_name', 'mask_path']
    imgs_info = []      # img_name, img_path, img_label
    # for img_name in os.listdir(cfg.tampered_img_paths):
    for img_dir in sorted(os.listdir(path)):
        print(img_dir)
        for img_name in sorted(os.listdir(os.path.join(path + img_dir))):
            print(img_name)
            if img_name.endswith('.png'):
                imgs_info.append([img_dir, img_name, os.path.join(path, img_name)])
                count += 1
    
    print(count)
    
    imgs_info_array = np.array(imgs_info)
    df = pd.DataFrame(imgs_info_array, columns=col_name)
    df.to_csv('./0327_mask.csv')
```

```python
import json
# js_file = f'/home/project/AMP_mysef_2D_Cls/MyDataset/name_to_T.json'
js_file = f'/home/workspace/AMP_mysef_2D_Cls/MyDataset/name_to_T.json'
def read_json_file(file_path):
    try:
        # Open the file and load the JSON data into a Python object
        with open(file_path, 'r') as file:
            json_data = json.load(file)
        return json_data
    except FileNotFoundError:
        # Raise an error if the file path does not exist
        raise FileNotFoundError(f"File not found: {file_path}")
    except ValueError:
        # Raise an error if the file does not contain valid JSON data
        raise ValueError(f"Invalid JSON data in file: {file_path}")   

name_to_Tstage = read_json_file(js_file)
```
