```python
import pandas as pd
```

```python
csv = pd.read_csv('./tstage.csv')
csv
```

```python
dict = {}

for i in range(len(csv)):
    # print(i)
    dict[csv['mri'][i]] = int(csv['T_stage'][i])
```

```python
dict
```

```python
import json
a = json.dumps(dict)
f2 = open('name_to_T.json', 'w')
f2.write(a)
f2.close()
```
