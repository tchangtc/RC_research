```python
padding_size = [128, 128, 128]
cropping_size = [128, 128, 128]

image_input = f'/home/workspace/AMP_mysef_3D_Cls/data/center_crop3d/imgTr/'
mask_input = f'/home/workspace/AMP_mysef_3D_Cls/data/center_crop3d/maskTr/'

image_out = f'/home/workspace/AMP_mysef_3D_Cls/data/final_out/imgTr/'
mask_out = f'/home/workspace/AMP_mysef_3D_Cls/data/final_out/maskTr/'

image_files = sorted(glob.glob(os.path.join(image_input + '*.nii.gz')))
mask_files  = sorted(glob.glob(os.path.join(mask_input + '*.nii.gz')))

count = 0
for data, seg in zip(image_files, mask_files):
    # print(image, label)
    tmp = data.split('/')[-1].split('.')[0]
    # print(tmp)
    name = tmp

    data = sitk.ReadImage(data)
    seg = sitk.ReadImage(seg)

    assert data.GetSpacing() == seg.GetSpacing()
    assert data.GetSize() == seg.GetSize()
    assert data.GetOrigin() == seg.GetOrigin()
    assert data.GetDirection() == seg.GetDirection()

    # print(data.GetSpacing())
    spacing = data.GetSpacing()
    print(round(spacing[0],4), round(spacing[1],4), round(spacing[2],4))
    # print(data.GetSize())
    # print(patient_name)

    info_dict['Spacing']    = data.GetSpacing()
    info_dict['Size']       = data.GetSize()
    info_dict['Origin']     = data.GetOrigin()
    info_dict['Direction']  = data.GetDirection()

    data_array, seg_array = sitk.GetArrayFromImage(data), sitk.GetArrayFromImage(seg)    
    assert data_array.shape == seg_array.shape
    # print(data_array.shape)

    z, y, x = data_array.shape
    
    img, lab = data_array, seg_array
    # pad if the image size is smaller than trainig size
    if z < padding_size[0]:
        if z % 2 == 0:
            # diff = (training_size[0]+2 - z) // 2
            diff = (padding_size[0] - z) // 2
            img = np.pad(img, ((diff, diff), (0,0), (0,0)))
            lab = np.pad(lab, ((diff, diff), (0,0), (0,0)))

        else:
            diff = (padding_size[0] - z) // 2
            img = np.pad(img, ((diff, diff+1), (0,0), (0,0)))
            lab = np.pad(lab, ((diff, diff+1), (0,0), (0,0)))
        
    if y < padding_size[1]:
        if y % 2 == 0:
            diff = (padding_size[1] - y) // 2
            img = np.pad(img, ((0,0), (diff, diff), (0,0)))
            lab = np.pad(lab, ((0,0), (diff, diff), (0,0)))
        
        else:
            diff = (padding_size[1] - y) // 2
            img = np.pad(img, ((0,0), (diff, diff+1), (0,0)))
            lab = np.pad(lab, ((0,0), (diff, diff+1), (0,0)))
        

    
    if x < padding_size[2]:
        if x % 2 == 0:
            diff = (padding_size[2] - x) // 2
            img = np.pad(img, ((0,0), (0,0), (diff, diff)))
            lab = np.pad(lab, ((0,0), (0,0), (diff, diff)))
        
        else:
            diff = (padding_size[2] - x) // 2
            img = np.pad(img, ((0,0), (0,0), (diff, diff+1)))
            lab = np.pad(lab, ((0,0), (0,0), (diff, diff+1)))
        



    if z > cropping_size[0]:
        size = cropping_size[0]
        img = img[z//2-size//2 : z//2+size//2, :, :]
        lab = lab[z//2-size//2 : z//2+size//2, :, :]

    if y > cropping_size[1]:
        size = cropping_size[1]
        img = img[:, y//2-size//2 : y//2+size//2, :]
        lab = lab[:, y//2-size//2 : y//2+size//2, :]

    if x > cropping_size[2]:
        size = cropping_size[2]
        img = img[:, :, x//2-size//2 : x//2+size//2]
        lab = lab[:, :, x//2-size//2 : x//2+size//2]

        


    # ResampleCRCMRImage(data, seg, image_out, mask_out, patient_name, count, (round(spacing[1],4), round(spacing[1],4), round(spacing[1],4)))
    print(img.shape, lab.shape)
    assert img.shape == lab.shape
    print(img.shape)
    count += 1
    data = sitk.GetImageFromArray(img)
    data.SetSpacing(info_dict['Spacing'])


    seg = sitk.GetImageFromArray(lab)
    seg.SetSpacing(info_dict['Spacing'])

    os.makedirs('%s' % (image_out), exist_ok=True)
    os.makedirs('%s' % (mask_out), exist_ok=True)
    # sample_image = sitk.WriteImage(data, f'/home/workspace/AMP_mysef_3D_Cls/data/crop3d/sample_image2.nii.gz')
    # sample_mask  = sitk.WriteImage(seg, f'/home/workspace/AMP_mysef_3D_Cls/data/crop3d/sample_seg2.nii.gz')
    sitk.WriteImage(data, '%s/%s.nii.gz' % (image_out, name))
    sitk.WriteImage(seg, '%s/%s.nii.gz' % (mask_out, name))
    print(name)
    # sitk.WriteImage(
    # print(name, 'done')

print(count)

```
